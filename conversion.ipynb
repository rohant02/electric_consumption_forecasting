```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('main_data.csv')
```

```python
df['DateTime'] = pd.to_datetime(df['DateTime'])
```

```python
df
```

```python
df = df[['DateTime','Global_active_power']]
```

```python
df
```

```python
df_daily = df.resample('M', on='DateTime').mean()
df_daily
```

```python
df_daily.plot(y='Global_active_power', title='Daily Global Active Power', legend=False)
```

```python
from scipy.stats import boxcox

# Drop NaN values before applying Box-Cox transformation
df_daily_clean = df_daily.dropna()

# Calculate Box-Cox lambda
_, lambda_value = boxcox(df_daily_clean['Global_active_power'])
lambda_value
```

```python
df_daily.to_csv('daily_data.csv', index=False)
```
